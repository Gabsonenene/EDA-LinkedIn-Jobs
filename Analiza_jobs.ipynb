# Opis plików oraz kolumn:

## job_postings.csv: 
* job_id: The job ID as defined by LinkedIn (https://www.linkedin.com/jobs/view/ job_id )
* company_id: Identifier for the company associated with the job posting (maps to companies.csv)
* title: Job title.
* description: Job description.
* max_salary: Maximum salary
* med_salary: Median salary
* min_salary: Minimum salary
* pay_period: Pay period for salary (Hourly, Monthly, Yearly)
* formatted_work_type: Type of work (Fulltime, Parttime, Contract)
* location: Job location
* applies: Number of applications that have been submitted
* original_listed_time: Original time the job was listed
* remote_allowed: Whether job permits remote work
* views: Number of times the job posting has been viewed
* job_posting_url: URL to the job posting on a platform
* application_url: URL where applications can be submitted
* application_type: Type of application process (offsite, complex/simple onsite)
* expiry: Expiration date or time for the job listing
* closed_time: Time to close job listing
* formatted_experience_level: Job experience level (entry, associate, executive, etc)
* skills_desc: Description detailing required skills for job
* listed_time: Time when the job was listed
* posting_domain: Domain of the website with application
* sponsored: Whether the job listing is sponsored or promoted.
* work_type: Type of work associated with the job
* currency: Currency in which the salary is provided.
* compensation_type: Type of compensation for the job.

## benefits.csv
* job_id: The job ID
* type: Type of benefit provided (401K, Medical Insurance, etc)
* inferred: Whether the benefit was explicitly tagged or inferred through text by LinkedIn

## companies.csv:
* company_id: The company ID as defined by LinkedIn
* name: Company name
* description: Company description
* company_size: Company grouping based on number of employees (0 Smallest - 7 Largest)
* country: Country of company headquarters.
* state: State of company headquarters.
* city: City of company headquarters.
* zip_code: ZIP code of company's headquarters.
* address: Address of company's headquarters
* url: Link to company's LinkedIn page

## job_skills.csv:
* job_id: The job ID
* skill_abr: Abbreviations corresponding to the skills required in a particular job 

## company_industreis.csv:
* company_id: The company ID as defined by LinkedIn
* industry: Industries of companies 

## company_specialities.csv:
* company_id: The company ID as defined by LinkedIn
* speciality: Company specialties

## employee_counts.csv:
* company_id: The company ID
* employee_count: Number of employees at company
* follower_count: Number of company followers on LinkedIn
* time_recorded: Unix time of data collection

```python
# Teraz muszę zająć się znormalizowaniem salary, żeby odnosiły się do jednego okresu - np. do YEARLY
periods = {'MONTHLY': 12,
            'Missing': 1,
            'HOURLY': 40*52,
            'YEARLY': 1,
            'WEEKLY': 52,
            'ONCE': 1
}

# wprowadzam to do salary 
for column in ['max_salary', 'med_salary', 'min_salary']:
    jobs_full[column] = jobs_full.apply(
        lambda row: row[column] * periods[row['pay_period']] if row['pay_period'] in periods else row[column],
        axis=1
    )
```

### Zauważyłam, że w większości duplikaty wynikają z różnego zapisu "and" i "&". Dlatego chcę przefiltorwać je i zostawić tylko jedną wersję.

```python
# Szukam duplikatów w company_in
duplicates_mask = company_indust.duplicated(subset='company_id', keep=False)
duplicates = company_indust[duplicates_mask]
# Chcę zachować tylko duplikaty z 'and' i usunąć te ze znakiem '&'
ampersand_mask = duplicates['industry'].str.contains('&')
and_mask = duplicates['industry'].str.contains(' and ')
filtered_duplicates = duplicates[and_mask | ~ampersand_mask]
company_indust_cleaned = company_indust[~duplicates_mask] # usuwam duplikaty z company_indust
company_indust = pd.concat([company_indust_cleaned, filtered_duplicates]) # dodaje przefiltrowane duplikaty do company_indust 
company_indust
```

```python
company_indust.company_id.duplicated().sum() # liczba duplikatów się zmniejszyła 
```

```python
company_indust = company_indust.drop_duplicates(subset='company_id') # usuwam pozostałe duplikaty
```

### Zauważyłam na wykresach, że problematyczne są wartości, które przekroczyły 100 M. Użyję IQR, żeby zidentyfikować wartości odstające. 

```python
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 10 * IQR
    upper_bound = Q3 + 10 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

max_salary_outliers_iqr = detect_outliers_iqr(salary_analyze['max_salary'])
min_salary_outliers_iqr = detect_outliers_iqr(salary_analyze['min_salary'])
med_salary_outliers_iqr = detect_outliers_iqr(salary_analyze['med_salary'])

all_outliers = max_salary_outliers_iqr | min_salary_outliers_iqr | med_salary_outliers_iqr
outliers_df = salary_analyze[all_outliers]
outliers_df
```

```python
# Usuwam te wartości, których min, max lub med salary są większe niż 10 M
condition = (salary_analyze['max_salary'] > 10_000_000) | \
            (salary_analyze['min_salary'] > 10_000_000) | \
            (salary_analyze['med_salary'] > 10_000_000)
salary_analyze = salary_analyze[~condition]
salary_analyze.reset_index(drop=True, inplace=True)
```
